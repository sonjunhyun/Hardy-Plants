```python
#기업 코드 padding해서 8자리 맞추기
def pad_code(code):
    code = str(code)
    if len(code) < 8:
        code = '0' * (8 - len(code)) + code
    return code
```

```python
#기업 코드 padding해서 8자리 맞추기
company_info['company_code'] = [pad_code(code) for code in company_info['company_code']]
```
