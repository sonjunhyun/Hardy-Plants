```python
#분석기업리스트 불러오기
import pandas as pd
companylist = pd.read_csv('./company_list3.csv', encoding='cp949')
companylist
```

```python
#코드 패딩
def pad_code(code):
    code = str(code)
    if len(code) < 6:
        code = '0' * (6 - len(code)) + code
    return code
```

```python
com_code = companylist['종목코드'].tolist()
com_code = [pad_code(code) for code in com_code]
companylist['종목코드'] = [pad_code(code) for code in companylist['종목코드']]
company = list(zip(companylist.iloc[:, 0], companylist.iloc[:, 1]))
```

```python
len(company)
```

```python
import pandas as pd
import requests


for code, company_name in company: 
    URL = f"https://finance.naver.com/item/main.nhn?code={code}"
    r = requests.get(URL)
    company_finance = pd.read_html(r.text)[3]
    company_finance = pd.DataFrame(company_finance)
    company_finance.to_csv(f'finance_{company_name}.csv', encoding='cp949')
    print(company_finance)

```
